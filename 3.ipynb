```python
import requests
import openpyxl
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment,Font

wrap_alignment = Alignment(wrap_text=True)
bold_font=Font(bold=True)
header_font=Font(bold=True,size=14)

url = 'https://davelee-fun.github.io/trial/board/news.html'
res= requests.get(url)
workbook = openpyxl.Workbook()
woorksheet = workbook.active
woorksheet.title='게시글'
woorksheet.append(['No','게시글 제목','댓글'])
woorksheet.column_dimensions['A'].width=20
woorksheet.column_dimensions['B'].width=100
woorksheet.column_dimensions['C'].width=100
soup = BeautifulSoup(res.content , 'html.parser')

data = soup.select('div.list_item.symph_row')
for cell in woorksheet[1]:
    cell.font = header_font
    
row= 2

for idx, i in enumerate(data[:5]):
    a= i.select_one('span.subject_fixed').get_text().strip()
    b= i.select_one('span.rSymph05').get_text().strip()
    c= i.select_one('a.list_subject')
    idx+=1
    link_url ='https://davelee-fun.github.io/trial/board/' + c['href']
    link = requests.get(link_url)
     
    soup_link = BeautifulSoup(link.content , 'html.parser')
    data_link = soup_link.find_all('div','comment_view')
    
    woorksheet.append([idx, a,'['+b+']'])
    woorksheet.cell(row=row, column=2).hyperlink=link_url

    for cell in woorksheet[row]:
        cell.alignment = wrap_alignment
        cell.font = bold_font
    row+=1
    
    for y in data_link:
        woorksheet.append(['','',y.get_text().strip().replace('\n','').replace(' ','')])
        for cell in woorksheet[row]:
            cell.alignment = wrap_alignment
        row+=1
   
workbook.save('inflearn.xlsx')
workbook.close()


```

```python
import requests
import re
from bs4 import BeautifulSoup
import openpyxl

workbook = openpyxl.Workbook()
worksheet = workbook.active
worksheet.title = '뉴진스'
worksheet.append(['그룹명','앨범명'])
worksheet.column_dimensions['B'].width = 200


url = 'https://www.yes24.com/Product/Search?domain=ALL&query=%EB%89%B4%EC%A7%84%EC%8A%A4&page='

for i in range(1,3):
    res = requests.get(url+str(i))
    soup = BeautifulSoup(res.content , 'html.parser')
    data = soup.select('div.info_row.info_name')

    for i in data:
        a = i.select_one('a.gd_name').get_text()
        if a[:3]=='뉴진스':
            b= re.sub('\[[^]]*\]','',a)
            c= re.sub('\([^)]*\)','',b)
            if '-' in c:
                name = c.split('-')[0]
                album = c.split('-')[1]
                worksheet.append([name, album])
                print(name,album)
                
workbook.save('Newjeans.xlsx')
workbook.close()

```
