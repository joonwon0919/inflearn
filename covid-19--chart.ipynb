```python
import pandas as pd
```

```python
import json, os
path = 'COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/'
```

```python
with open('COVID-19-master/csse_covid_19_data/country_convert.json', 'r', encoding='utf-8-sig') as json_file:
    json_data = json.load(json_file)
```

```python
def country_name_convert(row):
    if row['Country_Region'] in json_data:
        return json_data[row['Country_Region']]
    return row['Country_Region']

def create_dateframe(filename):
    
    doc = pd.read_csv(path + filename, encoding = 'utf8')
    try:
        doc = doc[['Country_Region','Deaths']]
    except:
        doc = doc[['Country/Region','Deaths']]
        doc.columns = ['Country_Region','Deaths']
    doc = doc.dropna(subset=['Deaths'])
    doc['Country_Region']  = doc.apply(country_name_convert, axis=1)
    doc =  doc.astype({'Deaths':'int'})
    doc = doc.groupby('Country_Region').sum()
    
    date_column = filename.split('.')[0].lstrip('0').replace('-','/')
    doc.columns = [date_column]
    
    return doc
    
        
```

```python
import os

def generate_dateframe(path):
    file_list, csv_list = os.listdir(path), list()
    
    first_doc = True
    for file in file_list:
        if file.split('.')[-1] =='csv':
            csv_list.append(file)
    csv_list.sort()
    
    for file in csv_list:
        doc = create_dateframe(file)
        if first_doc:
            final_doc, first_doc = doc , False
        else:
            final_doc = pd.merge(final_doc, doc, how='outer', left_index= True, right_index=True)
    final_doc= final_doc.fillna(0)
    return final_doc
```

```python

df_confirmed = generate_dateframe(path)
df_confirmed=df_confirmed.astype(int)

df_confirmed['Country_Flag']='https://s3.orbi.kr/data/file/united/e533d1c052be4ae83b34ddf643989ae0.jpg'

cols = df_confirmed.columns.tolist()
cols.remove('Country_Flag')
cols.insert(1,'Country_Flag')

df_confirmed=df_confirmed[cols]

df_confirmed.to_csv('COVID-19-master/final.csv')
```
